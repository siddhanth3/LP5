```python
# Import pandas for working with data tables.
# Import matplotlib for creating plots and visualizations.
import pandas as pd
import matplotlib.pyplot as plt
```

### Load Dataset

```python
# Read the Boston housing dataset from the CSV file.
# Display the first 10 rows to quickly understand the data.
df = pd.read_csv('Boston.csv')
df.head(10)
```

```python
# Remove extra unnamed columns that do not contain useful information.
df.drop(columns=['Unnamed: 15','Unnamed: 16'],inplace=True)
```

```python
# Remove the categorical MEDV column because we are predicting the numeric MEDV value.
df.drop(columns=['CAT. MEDV'],inplace=True)
```

#### Checking for null values

```python
# Check each column for missing values.
# A value of 0 means that column has no missing data.
df.isnull().sum()
```

```python
# Show column names, data types, and non-null counts.
# This helps confirm the dataset is clean and numeric.
df.info()
```

```python
# Show summary statistics such as mean, minimum, maximum, and quartiles.
# This gives a quick overview of the range of each feature.
df.describe()
```

#### Checking correlation with target variable MEDV

```python
# Calculate correlation of every feature with MEDV, the house price target.
# Sorting helps identify which features are most strongly related to price.
df.corr()['MEDV'].sort_values()
```

```python
# Select the input features used to predict house price.
# LSTAT, PTRATIO, and RM are chosen because they are strongly related to MEDV.
X = df.loc[:,['LSTAT','PTRATIO','RM']]

# Select MEDV as the target/output value to predict.
Y = df.loc[:,"MEDV"]

# Display the shapes to confirm the number of rows and columns.
X.shape,Y.shape
```

### Preparing training and testing data set

```python
# Import train_test_split to divide the data into training and testing sets.
from sklearn.model_selection import train_test_split

# Use 75% of the data for training and 25% for testing.
# random_state makes the split repeatable every time the code runs.
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.25,random_state=10)
```

### Normalizing training and testing dataset

```python
# Import StandardScaler to put features on a similar scale.
# Scaling helps neural networks train more smoothly.
from sklearn.preprocessing import StandardScaler
```

```python
# Create a scaler object that will learn the mean and standard deviation of the training data.
scaler = StandardScaler()
```

```python
# Fit the scaler only on the training data to avoid using information from the test data.
scaler.fit(x_train)
```

```python
# Transform both training and testing features using the same scaler.
# This keeps the feature values standardized for the neural network.
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

### Preparing model

```python
# Import Sequential for building the neural network layer by layer.
from keras.models import Sequential

# Import Dense to create fully connected neural network layers.
from keras.layers import Dense
```

```python
# Create an empty Sequential model.
# Layers will be added one after another.
model = Sequential()
```

```python
# Add the input layer with 128 neurons.
# input_shape=(3,) means each row has 3 input features.
model.add(Dense(128,input_shape=(3,),activation='relu',name='input'))

# Add a hidden layer with 64 neurons to learn more complex patterns.
model.add(Dense(64,activation='relu',name='layer_1'))

# Add the output layer with 1 neuron because we predict one price value.
# Linear activation is used for regression problems.
model.add(Dense(1,activation='linear',name='output'))

# Compile the model by choosing the optimizer, loss function, and evaluation metric.
# MSE is the loss, and MAE is easier to understand as average absolute error.
model.compile(optimizer='adam', loss='mse', metrics=['mae'])

# Display the model structure and number of trainable parameters.
model.summary()
```

```python
# Train the model using the training data.
# epochs=100 means the model sees the training data 100 times.
# validation_split=0.05 keeps 5% of the training data for validation during training.
model.fit(x_train,y_train,epochs=100,validation_split=0.05)
```

```python
# Evaluate the trained model on the unseen test data.
# This returns the loss value and the MAE metric.
output = model.evaluate(x_test,y_test)
```

```python
# Print the final test errors in a readable format.
# Lower MSE and MAE values mean better predictions.
print(f"Mean Squared Error: {output[0]}"
      ,f"Mean Absolute Error: {output[1]}",sep="\n")
```

```python
# Use the trained model to predict house prices for the test inputs.
y_pred = model.predict(x=x_test)
```

```python
# Print predicted prices beside the actual test prices for comparison.
print(*zip(y_pred,y_test))
```
